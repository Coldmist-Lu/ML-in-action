```python
from tensorflow import keras

(X_train_full, y_train_full), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()
X_train_full = X_train_full / 255.0
X_test = X_test / 255.0
X_valid, X_train = X_train_full[:5000], X_train_full[5000:]
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]

pixel_means = X_train.mean(axis=0, keepdims=True)
pixel_stds = X_train.std(axis=0, keepdims=True)
X_train_scaled = (X_train - pixel_means) / pixel_stds
X_valid_scaled = (X_valid - pixel_means) / pixel_stds
X_test_scaled = (X_test - pixel_means) / pixel_stds
```

## 1 Faster Optimizer

梯度下降：
$$\theta \leftarrow \theta - \eta\nabla_{\theta}J(\theta)$$

### 1-1 Momentum Optimizer 动量优化

* 主要思想：增加动量向量，以模拟摩擦机制。动量参数 $\beta$ 设置在0到1之间，0表示高摩擦，1表示无摩擦
* 缺点：增加了一个超参数
* 公式：
$$m \leftarrow \beta m - \eta\nabla_{\theta}J(\theta)$$
$$\theta \leftarrow \theta + m$$

```python
optimizer = keras.optimizers.SGD(lr=0.001, momentum=0.9)
```

### 1-2 Nesterov Accerlerated Gradient, NAG

* 对常规动量法进行改进，主要是并不继续推动越过谷底，而是稍微提前测量梯度，推回谷底。这样能够减少振荡。
* 通常来说，NAG 比常规的动量优化更快。
* 公式：
$$m \leftarrow \beta m - \eta\nabla_{\theta}J(\theta+\beta m)$$
$$\theta \leftarrow \theta + m$$

```python
optimizer = keras.optimizers.SGD(lr=0.001, momentum=0.9, nesterov=True)
```

### 1-3 AdaGrad

* 如果算法能更早的修正方向，更多指向全局最优。
* 缺点：有可能下降太快，往往停止得过早。因为学习率会按比例缩小。
* 公式：
$$s \leftarrow s + \nabla_{\theta}J(\theta) \otimes \nabla_{\theta}J(\theta)$$
$$\theta \leftarrow - \eta \oslash \sqrt{s+\varepsilon}$$

```python
optimizer = keras.optimizers.Adagrad(lr=0.001)
```

### 1-4 RMSProp

* 基于 AdaGrad 容易停止过早的问题，仅仅累加最近迭代中的梯度，通过指数衰减率来实现。
* 公式：
$$s \leftarrow \beta s + (1-\beta)\nabla_{\theta}J(\theta) \otimes \nabla_{\theta}J(\theta)$$
$$\theta \leftarrow - \eta \oslash \sqrt{s+\varepsilon}$$

```python
optimizer = keras.optimizers.RMSprop(lr=0.001, rho=0.9)
```

### 1-5 Adam & Nadam

* 是动量优化和 RMSProp 的结合
* 公式：
$$m \leftarrow \beta_1 m - (1-\beta_1)\nabla_{\theta}J(\theta)$$
$$s \leftarrow \beta_2 s + (1-\beta_2)\nabla_{\theta}J(\theta) \otimes \nabla_{\theta}J(\theta)$$
$$\hat{m} \leftarrow \frac{m}{1-\beta_1^t}$$
$$\hat{s} \leftarrow \frac{s}{1-\beta_2^t}$$
$$\theta \leftarrow - \eta \oslash \sqrt{s+\varepsilon}$$

* 式子中，t表示迭代次数，所以第3、4步主要是用来控制初期训练中 m、s 偏向0的情况
* $\beta_1 = 0.9, \beta_2 = 0.999$ (default)

```python
optimizer = keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999)
```

### 1-6 AdaMax

* 将 s 的 L2 范数替换为 L无穷范数。比 Adam 更加稳定（在某些数据集上）。

```python
optimizer = keras.optimizers.Adamax(lr=0.001, beta_1=0.9, beta_2=0.999)
```

### 1-7 Nadam

* Adam 优化加上 Nesterov 技巧

```python
optimizer = keras.optimizers.Nadam(lr=0.001, beta_1=0.9, beta_2=0.999)
```

## 2 Learning Rate Scheduling 学习率调度

### 2-1 Power Scheduling 幂调度

* 设置为迭代次数 t 的函数。$\eta(t) = \eta_0 / (1+t/s)^c$
* 每过 s 个步骤下降到 $\eta_0 / 2$, $\eta_0 / 3$, $\eta_0 / 4$, ...
* keras 中 c=1, decay=1/s

```python
optimizer = keras.optimizers.SGD(lr=0.01, decay=1e-4)
```

### 2-2 Exponential Scheduling 指数调度

* $\eta(t) = \eta_0 0.1^{t/s}$
* 每 s 步将下降 10 倍。


```python
# 设定方法 1 提供 lr0 和 s 参数
def exponential_decay(lr0, s):
    def exponential_decay_fn(epoch):
        return lr0 * 0.1**(epoch / s)
    return exponential_decay_fn

exponential_decay_fn = exponential_decay(lr0=0.01, s=20)
```

```python
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(300, activation="selu", kernel_initializer="lecun_normal"),
    keras.layers.Dense(100, activation="selu", kernel_initializer="lecun_normal"),
    keras.layers.Dense(10, activation="softmax")
])
model.compile(loss="sparse_categorical_crossentropy", optimizer="nadam", metrics=["accuracy"])
n_epochs = 25
```

```python
lr_scheduler = keras.callbacks.LearningRateScheduler(exponential_decay_fn)
history = model.fit(X_train_scaled, y_train, epochs=n_epochs,
                    validation_data=(X_valid_scaled, y_valid),
                    callbacks=[lr_scheduler])
```

```python
def exponential_decay_fn(epoch, lr): # 每次从以前的学习率开始。这和上一种是一样的，但是轮次从1就开始衰减。
    return lr * 0.1**(1 / 20) # 这边 epoch 参数虽然没用到，我个人认为是由于要送入回调，必须要配上这个参数。否则没法添加 lr
```

* 请注意，epoch 参数在训练中断后不会被保存，下一次训练会从 0 开始。由于学习率受epoch参数影响，因此可能导致很高的权重。
* 一个解决办法是手动设置 fit() 方法的 initial_epoch 参数，使 epoch 从正确的值开始。

* 如果我们希望 iteration 之间修改 learning rate 该怎么办呢？ 这里我们需要自定义一个回调：

```python
K = keras.backend

class ExponentialDecay(keras.callbacks.Callback):
    def __init__(self, s=40000):
        super().__init__()
        self.s = s
        
    def on_batch_begin(self, batch, logs=None): # 在每个 iteration 开始
        # the batch argument is reset at each epoch
        lr = K.get_value(self.model.optimizer.lr) # 取出 lr
        K.set_value(self.model.optimizer.lr, lr * 0.1 **(1/s)) # 调整 lr
        
    def on_epoch_end(self, epoch, logs=None): # 在每个 epoch 结束
        logs = logs or {}
        logs['lr'] = K.get_value(self.model.optimizer.lr) # 记录当前的 lr
        
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(300, activation="selu", kernel_initializer="lecun_normal"),
    keras.layers.Dense(100, activation="selu", kernel_initializer="lecun_normal"),
    keras.layers.Dense(10, activation="softmax")
])

lr0 = 0.01
optimizer = keras.optimizers.Nadam(lr=lr0)
model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
n_epochs = 25

s = 20 * len(X_train) // 32

exp_decay = ExponentialDecay(s)
history = model.fit(X_train_scaled, y_train, epochs=n_epochs,
                    validation_data=(X_valid_scaled, y_valid),
                    callbacks=[exp_decay])
```

### 2-3 Piecewise Constant Scheduling 分段恒定调度

* 对一些阶段用恒定的学习率。

```python
def piecewise_constant_fn(epoch):
    if epoch < 5:
        return 0.01
    elif epoch < 15:
        return 0.005
    else:
        return 0.001
```

```python
import numpy as np

def piecewise_constant(boundaries, values):
    boundaries = np.array([0] + boundaries) # [0, 5, 15]
    values = np.array(values)
    def piecewise_constant_fn(epoch):
        return values[np.argmax(boundaries > epoch) - 1]
    return piecewise_constant_fn

piecewise_constant_fn = piecewise_constant([5, 15], [0.01, 0.005, 0.001])

lr_scheduler = keras.callbacks.LearningRateScheduler(piecewise_constant_fn)

model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(300, activation="selu", kernel_initializer="lecun_normal"),
    keras.layers.Dense(100, activation="selu", kernel_initializer="lecun_normal"),
    keras.layers.Dense(10, activation="softmax")
])
model.compile(loss="sparse_categorical_crossentropy", optimizer="nadam", metrics=["accuracy"])
n_epochs = 25
history = model.fit(X_train_scaled, y_train, epochs=n_epochs,
                    validation_data=(X_valid_scaled, y_valid),
                    callbacks=[lr_scheduler])
```

### 2-4 Performance Scheduling 性能调度

* 使用 ReduceLROnPleateau 回调函数，每 5 个轮次都没有改善时就将学习率减半。

```python
import tensorflow as tf

tf.random.set_seed(42)
np.random.seed(42)
```

```python
lr_scheduler = keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=5)

model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(300, activation="selu", kernel_initializer="lecun_normal"),
    keras.layers.Dense(100, activation="selu", kernel_initializer="lecun_normal"),
    keras.layers.Dense(10, activation="softmax")
])
optimizer = keras.optimizers.SGD(lr=0.02, momentum=0.9)
model.compile(loss="sparse_categorical_crossentropy", optimizer=optimizer, metrics=["accuracy"])
n_epochs = 25
history = model.fit(X_train_scaled, y_train, epochs=n_epochs,
                    validation_data=(X_valid_scaled, y_valid),
                    callbacks=[lr_scheduler])
```

### 2-6 OneCycle Scheduling

* 周期调度的思想是，在训练前半部分先将学习率从 $\eta_0$ 线性增长到 $\eta_1$，最后再降低至 $\eta_0$。

```python
class OneCycleScheduler(keras.callbacks.Callback):
    def __init__(self, iterations, max_rate, start_rate=None,
                 last_iterations=None, last_rate=None):
        self.iterations = iterations
        self.max_rate = max_rate
        self.start_rate = start_rate or max_rate / 10
        self.last_iterations = last_iterations or iterations // 10 + 1
        self.half_iteration = (iterations - self.last_iterations) // 2
        self.last_rate = last_rate or self.start_rate / 1000
        self.iteration = 0
    def _interpolate(self, iter1, iter2, rate1, rate2):
        return ((rate2 - rate1) * (self.iteration - iter1)
                / (iter2 - iter1) + rate1)
    def on_batch_begin(self, batch, logs):
        if self.iteration < self.half_iteration:
            rate = self._interpolate(0, self.half_iteration, self.start_rate, self.max_rate)
        elif self.iteration < 2 * self.half_iteration:
            rate = self._interpolate(self.half_iteration, 2 * self.half_iteration,
                                     self.max_rate, self.start_rate)
        else:
            rate = self._interpolate(2 * self.half_iteration, self.iterations,
                                     self.start_rate, self.last_rate)
        self.iteration += 1
        K.set_value(self.model.optimizer.lr, rate)
```
