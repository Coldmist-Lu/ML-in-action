## Voting Classifier 投票分类

* 聚合各个分类器的预测结果，将得票最多的作为预测类别
* 弱学习器 --> 强学习器

### Lead-in: 大数定律

* 投掷一枚正面朝上概率 51% 的硬币 10000 次，统计正面朝上的概率

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
heads_proba = 0.51
coin_tosses = (np.random.rand(10000, 10) < heads_proba).astype(np.int32)
cumulate_heads = np.cumsum(coin_tosses, axis=0) / np.arange(1, 10001).reshape(-1, 1)
```

```python
plt.figure(figsize=(8, 3.5))
plt.plot(cumulate_heads)
plt.plot([0, 10000], [0.51, 0.51], 'k--', lw=2, label='51%')
plt.plot([0, 10000], [0.50, 0.50], 'k-', lw=2, label='50%')
plt.xlabel('Number of coin tosses')
plt.ylabel('Heads ratio')
plt.legend(loc='lower right')
plt.axis([0, 10000, 0.42, 0.58])
```

### Hard Voting 硬投票

* 最基本的投票方法，即对所有弱分类器取加权平均，得票最多为结果

```python
# 准备登月数据
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
lr_clf = LogisticRegression()
rf_clf = RandomForestClassifier()
svm_clf = SVC()
```

```python
voting_clf = VotingClassifier(
    estimators=[('lr', lr_clf), ('rf', rf_clf), ('svc', svm_clf)],
    voting='hard',
)
voting_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
for clf in (lr_clf, rf_clf, svm_clf, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

### Soft Voting 软投票

* 当分类器能够估算类别概率时，可根据各分类器的概率高低给予不同的权重。
* 通常比硬投票效果更好

```python
svc_prob_clf = SVC(gamma="scale", probability=True)
svoting_clf = VotingClassifier(
    estimators=[('lr', lr_clf), ('rf', rf_clf), ('svc', svc_prob_clf)],
    voting='soft',
)
svoting_clf.fit(X_train, y_train)
```

```python
for clf in (lr_clf, rf_clf, svm_clf, svoting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```
