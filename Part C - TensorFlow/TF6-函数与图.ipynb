# Tensorflow 6 - 函数与图

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

import os
os.environ['CUDA_VISIBLE_DEVICES'] = '2'
```

```python
def cube(x):
    return x ** 3
```

* 使用 Python 值来调用函数：

```python
cube(2)
```

* 使用张量来调用函数：

```python
cube(tf.constant(2.0))
```

* 现在将其转化成 TensorFlow 函数：

```python
tf_cube = tf.function(cube)
tf_cube
```

```python
tf_cube(2) # 返回的将是张量
```

```python
tf_cube(tf.constant(2.0))
```

* 可以使用 @tf.function 修饰：

```python
@tf.function
def tf_cube(x):
    return x ** 3

tf_cube(2)
```

```python
tf_cube.python_function(2)
```

* TensorFlow 会优化计算图，因此其计算速度通常会比 python 函数更快。
* Keras 会自动将你的自定义函数转化成 TF 函数 ———— 不需要使用 tf.function。
* 可以通过设置 dynamic=True 可以不将 Python 函数转化成 tf 函数。或者在模型 compile 时设置 run_eagerly=True。
